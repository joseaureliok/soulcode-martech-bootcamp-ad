## **NOÇÕES GERAIS DE ESTATÍSTICA**

### **DEFINIÇÃO**

Em síntese, a Estatística é a ciência que faz parta da Matemática Aplicada cujo objeto científico é **a coleta, a organização, descrição ou apresentação, análise e interpretação de dados** de uma população ou amostra.

Entende-se como **população** a totalidade do conjunto de dados analisado e **amostra**, por sua vez, pela parte destacada de uma população em que os elementos representem o todo (população).

### **CLASSIFICAÇÃO**

Pode-se dizer que a estatística está dividida em 03 (três) grandes grupos:
* **Probabilidade**
  
  > Preocupa-se com fenômenos aleatórios decorrentes de incerteza.

* **Estatística Inferencial**

  > Também conhecida como estatística indutiva, tem como foco análises e interpretações sobre uma amostra ou subconjunto da população para formulação de hipóteses e conclusões sobre aquela população.

* **Estatística Descritiva**

  > Utilizada para facilitar o entendimento (com gráficos ou não), produzir relatórios ou discussões, sintetiza e descreve dados numéricos, decorrentes do procedimento de coleta (elaboração), organização (tabulação e análise) e descrição (interpretção e apresetação) dos dados de um população ou amostra.
  
  > *É o foco desse curso*.

### **OUTROS CONCEITOS E DEFINIÇÕES IMPORTANTES**

#### **POPULAÇÃO E AMOSTRA**

Em síntese, são conjuntos de elementos ou indivíduos cujas características, propriedades e/ou atributos estão sob investigação estatística.

* **POPULAÇÃO**

  > **TODOS** os elementos ou indivíduos com as características, propriedades e/ou atributos investigados. Também é chamado de Universo.

* **AMOSTRA**

  > **SUBCONJUNTO FINITO** de uma população utilizada para obter as características daquela população.
  
  > *A amostra deve ser representativa, de modo que suas características, propriedades e/ou atributos sejam as mesmas ou as mais próximas possíveis da população*.  

  1. **Amostragem Aleaória Simples**

  > **TODOS** os elementos da população tem a mesma chance de serem selecionados.

  2. **Amostragem Estratificada**

  > A população é dividida em ao menos 2 (dois) estratos ou subpopulações com mesmas características, propriedades e/ou atributos, logo em seguida são selecionadas uma amostra de cada.

  3. **Amostragem Sistemática**

  > É selecionado um elemento ou indivíduo inicial, com características, propriedades e/ou atributos, a partir do qual são selecionados os demais de forma sistemática, ou seja, numa ordem *n* estabelecida.

  4. **Amostragem por Conglomerados**

  > Os elementos ou indivíduos são divididos em áreas ou conglomerados, as quais serão selecionadas por sorteio e analisadas em sua a integralidade.  


#### **VARIÁVEL**

De forma resumida, pode-se dizer que é a característica dos elementos ou indivíduos de uma amostra.

Pode ser:

* **QUALITATIVA**

  > Resulta de uma classificação por tipos ou atributos, geralmente não numéricos.

  Divide-se em:

  1. **NOMINAL**

  > Não é possível classificar ou sopesar  características, propriedades e/ou atributos sobre as outras.

  2. **ORDINAL**

  > Admite ordem aos elementos ou indivíduos em razão das características, propriedades e/ou atributos.

* **QUANTITATIVA**

  > Valores expressos em números.

  Diviem-se em:

  1. **DISCRETAS**

    > Obtidas de um conjunto enumerável por alguma forma de contagem.

  2. **CONTÍNUAS**

    > A princípio, são aquelas que podem assumir qualquer valor em determminado intervalo, geralmente, resultantes de medição e expressas em alguma unidade. Pertencem a um intevalor de números reais.

#### **COLETA, CRÍTICA E APRESENTAÇÃO DOS DADOS**

* **COLETA**

  > É a etapa seguinte a definição do problema de negócio ou objeto de estudo e análise de sua viabilidade, verificada pela busca ou compilação dos dados das variáveis selecionadas com as características, propriedades e/ou atribitos inicialmente delimitados.

  Pode ser de duas formas:

  1. DIRETA ou de FONTE PRIMÁRIA

  > Derivam da fonte originária. Ex: *datas de nascimentos, casamentos, e óbitos; documentos fiscais; etc*.

  2. INDIRETA ou de FONTE SECUNDÁRIA

  > São obtidas de outras fontes diretas, notadamente, por meio de interpretação. Ex: *cálculo de salário médio; variação do tempo de renteção de funcionário; etc*.

  Podem ser classificadas quanto ao tempo:

  1. **CONTÍNUA**: caráter permanente.
  
  2. **PERIÓDICA**: realizada em intervalos de tempo.

  3.  **OCASIONAL**: feita a qualquer tempo, sem qualquer previsão temporal.

* **CRÍTICA**

> É a limpeza dos dados e a consquente análise exploratória, selecionando dados integros e aptos a produzir informações relevantes.

* **APRESENTAÇÃO**

> Organização dos dados de maneira sintética, sistemática e racional, em **gráficos e/ou tabelas**.

> Adicionando ou subtraindo valores de todos os valore sou elementos de unma variável, a média será influenciada na mesma operação.

A **Amplitute Total** pode ser obtida pela subtração:

  > **MAX-MIN**

Já a **Amplitude Interquartílica** é obtida da subtração:

  > **Q3 - Q1**

#### **OUTLIERS, LIMITE INFERIOR (LI) E LIMITE SUPERIOR (LS)**

O **Limite Inferior** é obtido pela equação:

  > **Q1 - 1,5 * AIQ**

E **Limite Superior** pela equação:

  > **Q3 + 1,5 * AIQ**

Serão considerados **outliers** os elementos que:

  > forem maiores que o Limite Superior
    
    Elemento > LS

  > ou forem menores que o Limite Inferior
  
    Elemento < LI


### **MEDIDAS DE DISPERSÃO**

#### **Conceito**

  >  É o afastamento de todos os valores de uma série em relação a média aritmética ou mediana.
  
  > De acordo com a grandeza dos afastamentos as séries estatísticas podem ser: homogêneas (fraca dispersão quando C v < 30%) e heterogêneas (forte dispersão quando C v ≥ 30%)

* xi : 5, 5, 5, 5, 5 ⇒ **dispersão nula**
* y i : 3, 4, 5, 6, 7 ⇒ **fraca dispersão**
* z i : 1, 2, 5, 8, 9 ⇒ **forte dispersão**

  >


#### **Medidas de dispersão Absoluta | Relativa**

* **Absoluta**: dimensiona o afastamento médio entre os termos de uma série e seu valor médio. Seu resultado é expresso na mesma unidade de medida dos dados pesquisados.
* **Relativa**: Mede a relação entre a dispersão absoluta e o valor médio da série, com resultado expresso em porcentagens. Compara a dispersão de
universos diferentes (em unidade de medida e/ou valor médio), expressos em percentagens entre a relação das medidas de dispersão absoluta e de tendência central. São medidas usadas para análises comparativas.

**ASSIMETRIA**: É o grau de desvio ou afastamento da simetria de uma distribuição.
Quando a curva é simétrica, a média, a mediana e a moda coincidem,
num mesmo ponto, de ordenada máxima, havendo um perfeito equilíbrio
na distribuição. Quando o equilíbrio não acontece, isto é, a média, a mediana e a moda recaem em pontos diferentes da distribuição esta será
assimétrica; enviesada a direita ou esquerda.

* DISTRIBUIÇÃO SIMÉTRICA, ASSIMETRIA (S) = 0
  
  > **Moda = Mediana = Média**

      Q3 - Mediana = Q1 - Mediana

* DISTRIBUIÇÃO NEGATIVA
  > **Moda > Mediana > Média**

* ASSIMETRIA POSITIVA
  > **Moda < Mediana < Média**

#### **Critério de Bowley**

* **Assimetria Absoluta**
      Q3 - Mediana != Q1 - Mediana

      Q3 + Q1 - (2*Mediana) != 0

* **Assimetria Relativa**
  
      S = [Q3 + Q1-(2*Mediana)]/[Q3 - Q1], para -1<= S <= +1

    > S = -1 : negativo

      S = [-Q3 + Q1]/[Q3 - Q1]
    
    > S = +1 : positivo

      S = [Q3 - Q1]/[Q3 - Q1]


#### **Critério Kelley**

Para corrigir parte do inconveniente de se desprezar 50% das ocorrências (Critério de Bowley), Kelley aconselha o uso dos “Centis” eqüidistantes da mediana, tais como C10 e C90, para cálculo da assimetria.

    S = [C90 + C10 -(2*Mediana)]/[C90 - C10], limites entre -1 e +1

#### **Coeficiente de Assimetria de Pearson**

Á medida que a distribuição deixa de ser simétrica, a média, a mediana e a moda vão se afastando, aumentando cada vez mais a diferença entre elas.

    a = [3(Média - Mediana)]/s
ou
    
    a = [Média - Mediana]/s

**Curtose**

É o grau de achatamento de uma distribuição, em relação a distribui-
ção normal. A curtose pode ser de três tipos:

    C = Dq/[C90-C10]

    C = [Q3-Q1]/[2(C90-C10)]

* Mesocúrtica - quando a distribuição é normal

      C = 0,263

* Leptocúrtica - quando a distribuição é mais pontiaguda que a normal

      C < 0,263

* Platicúrtica - quando a distribuição é mais achatada que a normal

      C > 0,263


#### **Coeficiente de Correlação de Pearson**

O coeficiente de correlação de Pearson (r) é fortemente
influenciado pela média da distribuição. Por esse motivo, um dos
pressupostos centrais para que essa medida seja adequadamente utilizada
é de que as observações obedeçam a uma distribuição normal.

A correlação de Pearson avalia a relação linear entre duas variáveis contínuas. Uma relação é linear quando a mudança em uma variável é associada a uma mudança proporcional na outra variável.

Por exemplo, você poderia usar uma correlação de Pearson para avaliar se aumentos na temperatura da instalação de produção estão associados a uma redução da espessura da cobertura de chocolate.

    Considerando "X" como variável explicativa ou independente e "Y" como variável dependente:

    r = covariância (XY)/[desvio(X)* desvio(Y)]

        covariância (XY) = Soma(XY) - [(Soma (X) * Soma (Y))/n]]
        desvio (X) = Soma(X²) - [(Soma X)²/n]
        desvio (Y) = Soma(Y²) - [(Soma Y)²/n]

    O "n" representa o numero de elemtnos da população; ems e tratando de amostra, a representação ideal é "n-1" (número de elementos da amostra)

#### **Coeficiente de Correlação de Spearman**

A correlação de Spearman avalia a relação monotônica entre duas variáveis **contínuas** ou **ordinais**.

Em uma *relação monotônica*, as variáveis tendem a mudar juntas mas não necessariamente a uma *taxa constante*.

    > O coeficiente de correlação de Spearman baseia-se nos valores classificados de cada variável, em vez de os dados brutos.

A correlação de Spearman é muito usada para avaliar relações envolvendo variáveis ordinais. Por exemplo, você poderia usar a correlação de Spearman para avaliar se a ordem na qual os funcionários executam um teste está relacionada ao número de meses de emprego.

Sempre é uma boa idéia examinar a relação entre variáveis com um gráfico de dispersão. Os coeficientes de correlação medem apenas relacionamentos lineares (Pearson) ou monotônicos (Spearman). Outras relações são possíveis.

#### **Coeficiente de Determinação**

O **R-quadrado** é uma medida estatística de quão próximos os dados estão da linha de regressão ajustada. Ele também é conhecido como o coeficiente de determinação ou o coeficiente de determinação múltipla para a regressão múltipla.

A definição do R-quadrado é bastante simples:

    é a porcentagem da variação da variável resposta que é explicada por um modelo linear.

Ou:

    R-quadrado = Variação explicada/Variação total

O R-quadrado está sempre entre 0 e 100%:

    0% indica que o modelo não explica nada da variabilidade dos dados de resposta ao redor de sua média.

    100% indica que o modelo explica toda a variabilidade dos dados de resposta ao redor de sua média.

Em geral, quanto maior o R-quadrado, melhor o modelo se ajusta aos seus dados. No entanto, existem condições importantes para esta diretriz.

*Limitações:*

* O R-quadrado não pode determinar se as estimativas e predições dos coeficientes são tendenciosas, e é por isso que você deve avaliar os gráficos de resíduos.

* O R-quadrado não indica se um modelo de regressão é adequado. É possível ter um valor baixo de R-quadrado para um bom modelo ou um valor alto de R-quadrado para um modelo que não se encaixa nos dados!

Embora o R-quadrado forneça uma estimativa da força do relacionamento entre seu modelo e a variável resposta, ele não fornece um teste de hipóteses formal para esse relacionamento.

#### **Regressão Linear Simples**

A regressão linear simples é uma espécie de modelo na estatística cujo objetivo é indicar qual será o comportamento de uma variável dependente (Y) como uma função que contenha uma ou mais variáveis independentes (X).

*Utilidade:*

* verificar se uma variável **tem relação** com outra;
* **previsão** de uma variável pelo valor de outra variável;
* modelos base para regressão linear **múltipla**.


    y' = α + β*x

        y': valor previsto ou estimado ou seja, y ou variável dependente sobre a qual será realizada previsão;
        x: variável independente;
        α: indica o valor de “y”, quando “x” for igual a zero (x=0);
        β: a variável que determina a inclinação da reta, pois, identifica uma relação grande ou pequena.

    α = média de y + β*(média de x)

    β = Sxx/Sxy -► desvio padrão X/ covariância XY -► (Soma(X²) - [(Soma X)²/n]) / (Soma(XY) - [(Soma (X) * Soma (Y))/n])

#### **MEDIDAS DE ESTATÍSTICA DESCRITIVA**

```python
# função
def eda(dataframe):
  ''' Obtenção de medidas estatísticas como:
  - desvio padrão e variância
  - 1º e 3º quartil
  - máximo e mínimo
  - amplitude total
  - amplitude interquartílica
  - limites superior e inferior
  '''

  # desvio padrão
  sd = list(dataframe.iloc[2])
  sd.pop(0)

  # primeiro quartil
  q1 = list(dataframe.iloc[4])
  q1.pop(0)

  # terceiro quartil
  q3 = list(dataframe.iloc[6])
  q3.pop(0)

  # mínimo
  min = list(dataframe.iloc[3])
  min.pop(0)

  # máximo
  max = list(dataframe.iloc[7])
  max.pop(0)

  # variância
  var = []

  # amplitude total
  at = []

  # amplitude interquartílica
  aiq = []

  # limite superior e inferior
  ls = []
  li = []

  #laço for para passar os elementos na lista, considerando o tamanho
  for i in range(len(sd)):
    var.append(sd[i]**2)
    at.append(max[i]-min[i])
    aiq.append(q3[i]-q1[i])
    ls.append(q3[i]+(1.5*aiq[i]))
    li.append(q1[i]-(1.5*aiq[i]))

  #inclusão do nome de cada medida estatística na posição 0 (início)
  var.insert(0, 'variância')
  at.insert(0, 'AT')
  aiq.insert(0, 'AIQ')
  ls.insert(0, 'LS')
  li.insert(0, 'LI')

  # gravando dados da lista na variável, conforme localização com .loc[n]
  dataframe.loc[9] = var
  dataframe.loc[10] = at
  dataframe.loc[11] = aiq
  dataframe.loc[12] = ls
  dataframe.loc[13] = li
```

*Instanciação da função*

```python
# aplicação da função
eda(medidas)

medidas
```

OBS: difere-se de "Operação", ou seja, aquilo que é feito rotineiramente.

De forma geral, um projeto possui as seguintes etapas:
  
  1- FASE INICIAL: *overview sobre os dados e a conformidade dos requisitos ou recomendação do negócio*;

  2- PLANEJAMENTO:  *estrutação dos passos e atividades, com detalhamento e visão geral sobre o que será executado*;

  3- EXECUÇÃO: *implementação do projeto*;

  4- ENCERRAMENTO: *entrega*.

### **Projetos em Análise de Dados**



1. Entender o problema

        Utilizar de técnicas adequadas para analisar o problema, isto é, encontrar o ponto de atenção e melhoria, promovendo uma adequada orientação à sua solução.
        
2. Coletar os dados

        Entendido o que se prentende, passa-se a análise dos dados, verificando quais dados existem e quais podem ser obtidos para a consecução dos objetivos traçados para solução do problema.

3. Processar os dados

        Uma vez coletados os dados, estes devem ser processados e disponibilizados para as etapas seguintes, de modo a serem apresentados da melhor maneira, sem ruído ou, melhor, com integridade, evitando-se a tomada de decisões equivocadas.

4. Explorar os dados

        Disponíveis os dados de forma íntegra, pode-se realizar a análise exploratória, obtendo-se insights, informações textuais, numéricas e gráficas, para o entendimento do problema e encaminhamento de uma solução eficiente, ou a mais eficiente, para cada caso.

5. Comunicar resultados

        Obtidos resultados relevantes, encontrand-se uma ou várias possíveis soluções, as conclusões devem ser apresentadas, de forma clara, visual e compreensível, para auxiliar na tomada de decisões.

6. *Feedback*

        Concluídas as etapas anteriores, importente a exposição de *feedback* sobre o processo de análise e a influênncia na tomada de decisões, apontado eventuais vícios, erros ou ruídos, em qualquer das etapas já realizadas, buscando a melhoria contínua dos processos e da tomada de decisões.
