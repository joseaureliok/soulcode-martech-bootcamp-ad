```python
# importando bibliotecas
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

import pandas as pd
```

```python
# criando autenticaçao para o Google Drive
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

```python
# conectando ao link do arquivo do Google Drive
link = 'https://drive.google.com/file/d/1CGPtN9hf2BMDxkEL5VUZIhjXdl7fETW_/view?usp=drive_link'

id = link.split("/")[-2]

downloaded = drive.CreateFile({'id':id})
downloaded.GetContentFile('dataset_projeto01.csv')
```

```python
# lendo base de dados
df_base = pd.read_csv('dataset_projeto01.csv')
```

```python
# exibindo base de dados
display(df_base)
```

```python
# exibindo os tipos de dados
df_base.dtypes
```

```python
# aplicando describe no dataset
display(df_base.describe(include='all'))
```

```python
# aplicando mediana no dataset
df_median = df_base.median()
display(df_median)
```

```python
# aplicando moda no dataset
df_mode = df_base.mode("index", numeric_only=True, dropna=True)
display(df_mode.loc[0].round(2))
```

```python
# encontrado a variância do conjunto de dados
df_variancia = df_base.var()
display(df_variancia.round(2))
```

```python
# encontrando desvio absoluto
df_desv_abs = df_base.mad()
display (df_desv_abs)
```

```python
# encontrando a covarância
df_covariancia = df_base.cov()
display (df_covariancia)
```

```python
# encontrando correlação
df_correlacao = df_base.corr()
df_correlacao
```

```python
# moda do atributo idade
df_mode = df_base.mode("index", numeric_only=True, dropna=True)
df_mode.loc[0][0]
```

```python
# criação de dataframa Salário Mensal e Desgaste
df_salm = df_base[["Salário Mensal", "Desgaste"]]
df_salm.mode()
```

```python
df_salm.describe(include = "all").round(2)
```

```python
# resultados absolutos sobre Desgates
df_salm.groupby(['Desgaste']).describe().round(2)
```

```python
df_salm.groupby(['Desgaste']).var().round(2)
```
